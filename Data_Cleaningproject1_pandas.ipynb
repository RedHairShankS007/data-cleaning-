# **GETTING START BY IMPORTING PANDAS**

```python
import pandas as pd

# STEP-1 CREATING DATAFRAME OR USING IT TO VIEW DATASET FROM CSV OR EXCEL FILE IN PYTHON NOTEBOOK
data_set1 = "/Customer Call List.xlsx"
df = pd.read_excel(data_set1)
df
```

# **STEP-2 CHECK FOR DUPLICATES**

```python
df = df.drop_duplicates() # this will automatically drop duplicate dataset if they exist
df
```

# **STEP-3 DROP ENTIRE COLUMN WHICH IS NOT FOR USE**

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Check the original data types
print("Original Data Types:\n", df.dtypes)

# Change the 'Phone_Number' column from object to str
df['Phone_Number'] = df['Phone_Number'].astype(str)

# Check the updated data types
print("\nUpdated Data Types:\n", df.dtypes)
```

# **STEP-4 TO MODIFY THE FORMAT OF LAST_NAME COLUMN INTO NORMAL FORMAT BY STRIPING NON ALPHABETICAL CHARACHTERS**

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

# **REPLACING NAN AND NA WITH BLANK **

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
df = df.replace('Na','')
df=df.fillna('')
df
```

# ** Droping null vlaues from phone number column**

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```
